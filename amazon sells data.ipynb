```python
df["category"]=df["category"].str.split("|").str[0]
df
```

```python
df.rating.unique()
```

```python
df.rating=df.rating.replace("|",pd.NA)
df.dropna(inplace=True)
```

```python
df.rating.unique()
```

```python
df.rating=df.rating.astype("float")
```

```python
#q1>>  the average rating for each product category is----

df.groupby("category")["rating"].median()
```

```python
df.rating_count.unique()
```

```python
df.rating_count=df.rating_count.str.replace(",","").astype(float)
df.rating_count
```

```python
#q2>> the top rating_count products by category
data2=df.groupby("category")["rating_count"].median()
data2
```

```python
df['discounted_price'] = df['discounted_price'].str.replace('₹', '').str.replace(',', '').astype(float)
df['actual_price'] = df['actual_price'].str.replace('₹', '').str.replace(',', '').astype(float)

plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.hist(df['discounted_price'], bins=10, color='skyblue', edgecolor='black')
plt.title('Distribution of Discounted Prices')
plt.xlabel('Discounted Price')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df['actual_price'], bins=10, color='salmon', edgecolor='black')
plt.title('Distribution of Actual Prices')
plt.xlabel('Actual Price')
plt.ylabel('Frequency')


# the distribution of discounted prices vs. actual prices
```
