```python
import pandas as pd
```

```python
df=pd.read_excel("Customer Call List.xlsx")
df
```

```python
df1=df.drop_duplicates()
df1
```

```python
df.drop(columns="Not_Useful_Column")
```

```python
df
```

```python
#by default strip uses for removing spaces from string value
#we clean the column = "Last_Name"
df["Last_Name"]=df["Last_Name"].str.lstrip("...") # ".." remove from data
df["Last_Name"]=df["Last_Name"].str.rstrip("...") #rstrip work from left side
df["Last_Name"]=df["Last_Name"].str.strip("/")
df["Last_Name"]=df["Last_Name"].str.rstrip("_") #remove "_" from right
df["Last_Name"]=df["Last_Name"].str.strip("123")#if number present it remove thoes number from that column
df
```

```python
#removing different format from Phone_number column, we want all of the data look exactly same

df["Phone_Number"]=df["Phone_Number"].str.replace(r"[^a-zA-Z0-9]", "",regex=True) #replace ^,a-z,A-Z,0-9 data, replace with " ", "^" means any charecter

df["Phone_Number"]=df["Phone_Number"].apply(lambda x: str(x)) #this line will convert cols data into string, so when we run next line it dont throw float error

df["Phone_Number"]=df["Phone_Number"].apply(lambda x: x[0:3]+"-" + x[3:6]+"-" + x[6:10])

df["Phone_Number"]=df["Phone_Number"].str.replace("nan--","") # remove "nan--"

df["Phone_Number"]=df["Phone_Number"].str.replace("Na--","")

df["Phone_Number"]=df["Phone_Number"].str.replace("--", "")

df
```

```python
# now we work with address column
df["Address"].str.split(",",n=2,expand=True) #in this line we split the address by "," upto 2 . then expand it
#then put thoes three different column data to new three column data "street address","state","zipcode".
```

```python
df[["street address","state","zipcode"]]=df["Address"].str.split(",",n=2,expand=True)
#then put thoes three different column data to new three column data "street address","state","zipcode".
df
```

```python
#now fro Paying Customer
#we want to convert no=n,yes=y

df["Paying Customer"]=df["Paying Customer"].str.replace("Yes","Y")
df["Paying Customer"]=df["Paying Customer"].str.replace("No","N")
df
```

```python
#now for  Do_Not_Contact

df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace("Yes","Y")
df["Do_Not_Contact"]=df["Do_Not_Contact"].str.replace("No","N")
df
```

```python
df=df.replace("N/a","") #as we apply in the whole data set, that contain string,float vales, thats why we dont use .str in this code
df
```

```python
# filling the Nan values, Nan value are not a number, dosnt exist so we cant use replace here

df.fillna("")
```

```python
# now we want to remove the thoes number arent callable, and do_not_contact contain N also

for x in df.index: #as we want it index wise operation
    if df.loc[x,"Do_Not_Contact"]=="N": # apply condition on column name Do_Not_Contact
         df.drop(x, inplace=True)

df
```

```python
for i in df.index:
    if df.loc[i,"Phone_Number"]=="":
        df.drop(i,inplace=True)

df
```

```python
df.drop(columns="Address",inplace=True)
df
```
