```python
#rename multiple columns for a better understanding

df=df.rename(columns={"NO":"NOT_OUT",
                      "HS":"HIGHEST_SCORE",
                     "BF":"BALL_FACED",
                     "SR":"STRIKE_RATES",
                     "MAT":"MATCHES"})
df
```

```python
#check null value

df.isnull().any()
```

```python
#as above code shows that STRIKE_RATES has a null value, to see that null value row
df[df["STRIKE_RATES"].isna()==True]
```

```python
#to fix it
df["STRIKE_RATES"]=df["STRIKE_RATES"].fillna(0)
df.isnull().sum() #so now no more null value is the dataframe
```

```python
df[df["Player"]=="ED Weekes (WI)"] #now see 0 filled inside the null value
```

```python
df=df.drop_duplicates()
df
```

```python
#split up span into start and end date

df[["starting_year","last_year"]]=df["Span"].str.split("-",expand=True)
df
```

```python
df=df.drop(columns="Span")
```

```python
#removing country code from the player column

df["Player"].str.split("(",expand=True)
```

```python
df["Player"].str.split("(").str[0]
#spliting the player col, and assigning the 1st index value as player_name
```

```python
df["Player_name"]=df["Player"].str.split("(").str[0]
df
```

```python
df["country"]=df["Player"].str.split("(").str[1]
df
```

```python
df=df.drop(columns="Player",axis=0)
df
```

```python
df.dtypes
```

```python
#now see the highest score is a object dtypes and we want to remove the *,and convert it into int

df['HIGHEST_SCORE']=df['HIGHEST_SCORE'].apply(lambda x: x.replace('*', '') if isinstance(x, str) else x)
#The lambda function checks if the element is a string (isinstance(x, str)). If it is, it replaces "*"
#with an empty string (x.replace('*', '')). If it's not a string (e.g., NaN), it leaves the element unchanged (else x).                          
df
```

#if any columns consist of this kind of issue, then we can do like above code

for i in df["HIGHEST_SCORE"]:

        print(i)

Check if the Entry is a String: The isinstance(i, str) check ensures that only string values are checked for the presence of "*". This prevents the error you encountered.        

```python
df["HIGHEST_SCORE"]=df["HIGHEST_SCORE"].astype("int")
df.dtypes
```

```python
#converting data types of columns in a one go 

df=df.astype({"starting_year":"int",
             "last_year":"int",
             })
df
```

```python
pd.set_option("display.max_rows",None)
df
```

```python
#now we removing + from the BALL_FACED

df.dropna() #droping nan values
df.replace("-","") # to replace "-", by ""
df.isnull().any() #to check if any null present 

df['BALL_FACED'] = df['BALL_FACED'].apply(lambda x: x.strip("+") if isinstance(x, str) else x)
#isinstance(x, str): Check if x is a string,x.strip("+"): Remove the "+" character from the string.

#df["BALL_FACED"].astype("int")
df
```

```python
#REMOVING + FROM THE 4S col
df["4s"]=df["4s"].apply(lambda x: x.strip("+") if isinstance(x,str) else x)
```

```python
#removing ) from the country 
df["country"]=df["country"].apply(lambda x: x.strip(")") if isinstance(x,str) else x)
```

```python
df
```

```python
#carrer lenght of any player

df["carrer_lenght"]=df["last_year"]-df["starting_year"]
df
```

```python
#avg career length of all players

df["carrer_lenght"].mean()
```

```python
#avg batting strike rate ,thoes who play more than 10 years
df[df["carrer_lenght"]>10]["STRIKE_RATES"].mean()
```

```python
#find number of player played before 1950
df[df["last_year"]<1950]
```

```python
#4 max highest score by country
df.groupby("country")["HIGHEST_SCORE"].max()
df.sort_values(ascending=False)
```
