```python
q1=df["km_driven"].quantile(0.25)
q3=df["km_driven"].quantile(0.75)
IQR=q3-q1

lower_limit=IQR-1.5*q1
upper_limit=IQR+1.5*q3
```

```python
df["km_driven"] = df["km_driven"].apply(lambda x: lower_limit if x < lower_limit else (upper_limit if x > upper_limit else x))

sns.boxplot(x=df["km_driven"])

#we replace the outliers value inside the df.km_driven columns
```

```python
import datetime
current_year = datetime.datetime.now().year
df['age'] = current_year - df['year']
df['depreciation'] = df['selling_price'] / df['age']
average_depreciation = df['depreciation'].mean()

print(f"The average depreciation in selling price based on the bike's age is {average_depreciation:.2f} per year.")
```
