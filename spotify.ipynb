```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

import warnings
warnings.filterwarnings("ignore")
```

```python
# the distribution of popularity among the tracks in the dataset.

sns.histplot(x=df["Popularity"])
plt.xticks(rotation=90)
plt.show()
```

```python
sns.scatterplot(y=df.Popularity,x=df["Duration (ms)"])

#most popular song are between 100000ms to 300000ms.
```

```python
artist_counts = df['Artist'].value_counts().head(10)
sns.countplot(data=df, y='Artist', order=artist_counts.index, palette='coolwarm')
plt.xticks(rotation=90)
plt.show()
```

```python
#top 5 least popular tracks in the dataset & the artist name and track name
df.sort_values(by="Popularity").head(5)
```

```python
#the top 5 most popular artists

df.sort_values(by="Popularity",ascending=False).head(5)
```

```python
# artist has the highest popularity on average

df.groupby("Artist")["Popularity"].median().sort_values(ascending=False)
```

```python
#calculate the median 'Popularity'
artist_median_popularity = df.groupby('Artist')['Popularity'].median().sort_values(ascending=False)
# top 5 most popular artists
top_artists = artist_median_popularity.head(5).index

# Get the most popular track for each of the top 5 artists
most_popular_tracks = df[df['Artist'].isin(top_artists)].sort_values(by=['Artist', 'Popularity'], ascending=[True, False])

print(most_popular_tracks[['Artist', 'Track Name', 'Popularity']].drop_duplicates('Artist'))
```

```python
top_10_artists = df['Artist'].value_counts().head(10).index
sns.violinplot(data=df[df['Artist'].isin(top_10_artists)], 
               y='Duration (ms)', 
               x='Artist',
               order=top_10_artists)
plt.xticks(rotation=90)
plt.show()
```

```python
#How does the distribution of track popularity vary for different artists? Visualize this using a swarm plot or a
#violin plot

top_10_artists = df['Artist'].value_counts().head(10).index
sns.violinplot(data=df[df['Artist'].isin(top_10_artists)], 
               x='Popularity', 
               y='Artist',
               order=top_10_artists,
               palette='coolwarm')

plt.title('Distribution of Track Popularity by Artist')
plt.xlabel('Popularity')
plt.ylabel('Artist')
plt.show()
```
